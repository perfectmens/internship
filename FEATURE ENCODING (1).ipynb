### Frequency Encoding: This method replaces each category with its frequency or the count of how many times it appears in the dataset.

```python
dt={"fruits":["apple","mango","banana","apple","banana","apple","mango","banana","chikku"]
}


df=pd.DataFrame(dt)
df
```

```python
frequency = df['fruits'].value_counts().to_dict()

'''value_counts(): This method counts the occurrences of each unique value in the 'nom_0' column.
The result is a Series where the index is the unique values from 'nom_0', 
and the values are the counts (frequencies) of those unique values.
This converts the Series to a dictionary. The keys of the dictionary are the unique values from 'nom_0', 
and the values are their corresponding frequencies.'''

df['encoded'] = df['fruits'].map(frequency)

print(df)
```

```python
from sklearn.preprocessing import OneHotEncoder

# Initialize the OneHotEncoder
encoder = OneHotEncoder(sparse=False)  #sparse_output=False    #when sparse=False, the result is a normal table (like a pandas DataFrame or a NumPy array) 
                                   #that’s easier to understand and manipulate.easier for humans to read,used in smaller datasets

# Apply OneHotEncoder to the 'nom_0' column
encoded_columns = encoder.fit_transform(df[['color']])

# Create a DataFrame for the encoded columns with proper column names
encoded_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(['color'])).astype(int)  # Convert to integer values (optional for simplicity)

# Drop the original 'nom_0' column from the original DataFrame
df = df.drop('color', axis=1)

# Concatenate the encoded columns with the remaining columns
df = pd.concat([df, encoded_df],axis=1)

# Display the resulting DataFrame
df
```

### ordinal encoding

### Ordinal Encoding is a technique used to convert categorical data into numeric data by assigning an ordered integer to each category.

```python
import pandas as pd
data = {
    'ord_2': ['Low', 'High', 'High', 'Medium', 'Low']
}

df = pd.DataFrame(data)
df
```

```python
import pandas as pd

# Sample data
data = {'values': [-1,-2,10, 12, 14, 15, 15, 16, 18, 19, 120]}
df = pd.DataFrame(data)

# Step 1: Calculate Q1 and Q3
Q1 = df['values'].quantile(0.25)
Q3 = df['values'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)

# Step 3: Filter out outliers
df_no_outliers = df[(df['values'] >= lower_bound) & (df['values'] <= upper_bound)]

print("Original Data:")
print(df)
print("\nAfter Removing Outliers:")
print(df_no_outliers)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


mu = 4        # Mean
sigma = 1     # Standard deviation
x_value = 4.25

# Z-score
z = (x_value - mu) / sigma
left_area = norm.cdf(z)
right_area = 1 - left_area

# Print the areas
print(f"Z = {z:.2f}")
print(f"Left area (P[X < {x_value}]) = {left_area:.4f}")
print(f"Right-tail area (P[X > {x_value}]) = {right_area:.4f}")

# Plotting
x = np.linspace(1, 7, 1000)
y = norm.pdf(x, mu, sigma)

plt.plot(x, y, label='Normal Distribution', color='black')
plt.fill_between(x, y, where=(x > x_value), color='red', alpha=0.4, label='Right Tail (Jail)')
plt.axvline(x_value, color='green', linestyle='--', label=f'x = {x_value}')
plt.axvline(mu, color='gray', linestyle='--', label=f'Mean (μ={mu})')

plt.text(4.05, 0.2, f'{left_area:.4f}', fontsize=10)
plt.text(5, 0.05, f'1 - Left Area = {right_area:.4f}', fontsize=10)

plt.title('Right-Tail Area Beyond Z = 0.25')
plt.xlabel('Score')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
