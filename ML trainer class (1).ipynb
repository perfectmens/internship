# **Mechine Learning**
 
 Machine Learning is a subset of artificial intelligence (Ai).
It makes the machine learn from past data and make predictions.
It involves training algorithms to learn patterns from data and make predictions or decisions.

### Work flow for mechine learning :-

1- **data collection** (gather data from various sources). 

2- **data preprocessing** (clean and prepare data analysis). 

3-**data analysis,data visualization.** (Exploring the data to understand its structure, patterns, and relationships.) 

4-**feature engneering.**(Creating new features or modifying existing ones to improve the model's performance.) 

5-**model training.**(Using the prepared data to train a machine learning model.) 

6-**model evaluation.**(Assessing the model's performance using metrics like accuracy, precision, recall, F1 score, and mean squared error.) 

7-**model deployment.**(Integrating the trained model into a production environment where it can make real-time predictions or decisions based on new data.)

### Mechine learning has two types supervised and un supervised learning
**Supervised:-** Regression and Classification

**Unsupervised:-** clustering

### Supervised and Unsupervised
"Supervised machine learning is trained using well-labeled training data. This means some input data is already tagged with the correct output. Finding the relationship between the input variable and the output variable is called regression. The input variable is known as the independent variable or feature (which can be any value), while the output is known as the dependent variable or target (there is always one dependent variable and one output)."

"Unsupervised machine learning involves training machines using unlabeled training data. Based on this data, the machine identifies patterns and structures without predicting specific outcomes. Clustering is a method used in unsupervised learning to group similar data points into clusters. The goal is to have data points within the same cluster that are more similar to each other than to those in other clusters."

## Regression: Regression algorithms are used to predict continuous outcomes based on input data.
1)**Linear Regression:** When there is a linear relationship between the dependent and independent variables.

**Applications:** Predicting house prices, sales forecasting, and risk assessment.

### Classification:  classification algorithms are used to predict categorical outcomes based on input data.
**1)Logistic Regression:** Models the probability of a binary outcome (0 or 1) based on one or more predictor variables.

*Applications:* Spam detection, disease diagnosis.

**2)k-Nearest Neighbors (k-NN):** Classifies a data point based on the majority class among its k nearest neighbors.

*Applications:* Recommender systems, anomaly detection.

**3)Decision Trees:** Splits the data into branches to form a tree based on feature values, making decisions at each node.

*Applications:* Customer segmentation, credit risk analysis.

**4)Random Forest:** An ensemble method that combines multiple decision trees to improve accuracy and reduce overfitting.

Feature selection, fraud detection.

**5)Support Vector Machines (SVM):** Finds the optimal hyperplane that separates classes in a high-dimensional space.

*Applications:* Image classification, text categorization.

**6)Naive Bayes:** Based on Bayes' theorem, assuming independence between features.

*Applications:* Text classification, spam filtering.

## Clustering
"K-means clustering is a method used to group similar data points together into clusters. The goal is to have data points in the same cluster that are more similar to each other than to those in other clusters. Here, we do not predict outcomes but rather find patterns. We do not use train-test splits; only independent variables are given."

```python
import numpy as np    
import matplotlib.pyplot as plt    
import pandas as pd    
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
data={
    'students':('a','b','c','d','e','f','g','h'),
    'maths score':(89,67,78,45,62,98,78,98),
    'science score':(89,78,90,78,67,89,98,78),
}
```

```python
k=3
```

```python
kmeans=KMeans(n_clusters=k,random_state=0)
kmeans.fit(x)
```

```python
labels=kmeans.labels_
df['cluster']=labels
df['cluster']
```

```python
df
```

```python
plt.figure(figsize=(10,6))
colors =plt.get_cmap('viridis',k)
for cluster in range(k):
    cluster_data=df[df['cluster']==cluster]
    plt.scatter(cluster_data['maths score'],cluster_data['science score'],color=colors(cluster),label=f'cluster{cluster+1}')
centroids=kmeans.cluster_centers_
plt.scatter(centroids[:,0],centroids[:,1],color='black',marker='x',s=100,label='centroids')#s is size centroid
plt.xlabel('maths score')
plt.ylabel('science score')
plt.title('maths score')
plt.legend()
plt.show
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score

# Sample Data (replace this with actual social media data)
data = {
    'tweet': [
        "I love the new design!",
        "This is the worst update ever.",
        "Amazing features in this app.",
        "Not happy with the service.",
        "I’m so excited to use this!",
        "Terrible experience, disappointed.",
        "Best app ever! So useful.",
        "Very bad UI, not user-friendly."
    ],
    'sentiment': [
        "positive", "negative", "positive", "negative",
        "positive", "negative", "positive", "negative"
    ]
}

# Load data into a DataFrame
df = pd.DataFrame(data)

# Convert sentiment to binary (positive = 1, negative = 0)
df['sentiment'] = df['sentiment'].map({'positive': 1, 'negative': 0})

# Split into training and testing data
X_train, X_test, y_train, y_test = train_test_split(df['tweet'], df['sentiment'], test_size=0.25, random_state=42)

# Vectorize text data
vectorizer = CountVectorizer()
X_train_vectors = vectorizer.fit_transform(X_train)
X_test_vectors = vectorizer.transform(X_test)

# Train a Logistic Regression model
model = LogisticRegression()
model.fit(X_train_vectors, y_train)

# Predict and evaluate
y_pred = model.predict(X_test_vectors)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))

# Test on new tweet
test_tweet = ["I hate the new update!"]
test_vector = vectorizer.transform(test_tweet)
prediction = model.predict(test_vector)
print("\nSentiment Prediction:", "Positive" if prediction[0] == 1 else "Negative")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression

# Sample DataFrame
data = pd.DataFrame({
    'age': [21, 35, 45, 18],
    'comment': ['I love it', 'Terrible service', 'Great experience', 'Not good'],
    'label': [1, 0, 1, 0]
})

# Feature & Label split
X = data[['age', 'comment']]
y = data['label']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Preprocessing: text + numeric
preprocessor = ColumnTransformer(
    transformers=[
        ('text', TfidfVectorizer(), 'comment'),     # Text column
        ('num', StandardScaler(), ['age'])          # Numeric column
    ]
)

# Create Pipeline
clf = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', LogisticRegression())
])

# Train model
clf.fit(X_train, y_train)

# Predict
predictions = clf.predict(X_test)
print(predictions)
```
