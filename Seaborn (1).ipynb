```python
# Check if a number is an Armstrong number
number = int(input("Enter a number: "))
sum_of_powers = 0
temp = number
x=str(number)
b=len(x)
#print(b)

# Calculate the sum of the cubes of the digits
while temp > 0:
    digit = temp % 10
    sum_of_powers += digit ** b
    temp //= 10

# Check if the sum equals the original number
if sum_of_powers == number:
    print(f"{number} is an Armstrong number.")
else:
    print(f"{number} is not an Armstrong number.")
```

```python
def find_missing_number(arr, n):
    # Calculate the sum of the first n natural numbers
    total_sum = n * (n + 1) // 2
    
    # Calculate the sum of the elements in the array
    array_sum = sum(arr)
    
    # The missing number is the difference between the total sum and the array sum
  
    return total_sum - array_sum

# Input array and the value of n
arr = [1, 2, 4, 5]
n = 5  # The largest number in the range

# Find and print the missing number
missing_number = find_missing_number(arr, n)
print(f"The missing number is: {missing_number}")
```

```python
# Step 1: Importing Required Libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
import matplotlib.pyplot as plt

# Step 2: Load the Dataset
iris = load_iris()
X = iris.data  # Features

# Step 3: K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
clusters_kmeans = kmeans.labels_
silhouette_kmeans = silhouette_score(X, clusters_kmeans)

# Step 4: Hierarchical Clustering
hierarchy = linkage(X, method='ward')
clusters_hierarchy = fcluster(hierarchy, t=3, criterion='maxclust')
silhouette_hierarchy = silhouette_score(X, clusters_hierarchy)

# Step 5: Plot the Accuracy (Silhouette Scores)
# Store algorithms and their scores
algorithms = ['K-Means', 'Hierarchical Clustering']
scores = [silhouette_kmeans, silhouette_hierarchy]

# Plot the bar chart
plt.figure(figsize=(8, 5))  # Set the figure size
plt.bar(algorithms, scores, color=['skyblue', 'orange'], edgecolor='black')  # Bar chart with colors
plt.ylim(0, 1)  # Set the y-axis range (Silhouette Score ranges between -1 to 1)
plt.title('Clustering Algorithm Comparison (Silhouette Scores)', fontsize=14)  # Title of the graph
plt.xlabel('Algorithms', fontsize=12)  # X-axis label
plt.ylabel('Silhouette Score', fontsize=12)  # Y-axis label
plt.xticks(fontsize=10)  # Set x-axis tick size
plt.yticks(fontsize=10)  # Set y-axis tick size
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for better readability
plt.show()  # Display the plot
```

```python
# Step 1: Import required libraries
import numpy as np
import pandas as pd
import seaborn as sns  # For advanced visualizations
import matplotlib.pyplot as plt  # For visualization

# Step 2: Prepare the data
algorithms = ['K-Means', 'Hierarchical Clustering']  # Names of algorithms
scores = [0.55, 0.57]  # Example silhouette scores
categories = ['Unsupervised', 'Unsupervised']  # Both algorithms belong to the same category

# Combine data into a DataFrame for Seaborn
data = pd.DataFrame({
    'Algorithms': algorithms,
    'Scores': scores,
    'Category': categories  # A column to apply 'hue'
})

# Step 3: Plot using Seaborn
plt.figure(figsize=(8, 5))  # Set the figure size
sns.barplot(x='Algorithms', y='Scores', hue='Category', data=data, palette='coolwarm')  # Bar plot with hue
plt.ylim(0, 1)  # Silhouette Score ranges between -1 and 1
plt.title('Clustering Algorithm Comparison (Silhouette Scores)', fontsize=14)  # Title of the graph
plt.xlabel('Algorithms', fontsize=12)  # X-axis label
plt.ylabel('Silhouette Score', fontsize=12)  # Y-axis label
plt.xticks(fontsize=10)  # Set x-axis tick size
plt.yticks(fontsize=10)  # Set y-axis tick size
plt.legend(title='Category', fontsize=10)  # Add a legend for the hue
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add a grid for better readability
plt.show()  # Display the plot
```
