```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
df=pd.read_csv('data.csv')
```

```python
df
```

```python
plt.figure(figsize=(8, 5))
plt.


plot(df["Duration"], df["Calories"], marker='o', linestyle='-', color='b')
plt.title("Line Plot: Duration vs Calories")
plt.xlabel("Duration (min)")
plt.ylabel("Calories")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df["Pulse"], df["Calories"], color='r', alpha=0.7)
plt.title("Scatter Plot: Pulse vs Calories")
plt.xlabel("Pulse")
plt.ylabel("Calories")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(df["Duration"][:10], df["Calories"][:10], color='g', alpha=0.6,width=1.5)
plt.title("Bar Chart: Duration vs Calories (First 10 Records)")
plt.xlabel("Duration (min)")
plt.ylabel("Calories")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
duration_counts = df["Duration"].value_counts().head(5)
plt.pie(duration_counts, labels=duration_counts.index, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'purple', 'orange'])
plt.title("Pie Chart: Top 5 Duration Frequencies")
plt.show()
```
